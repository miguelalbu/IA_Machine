```python
# Importando bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Configurando o estilo dos gráficos
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (12, 8)

# Carregando os dados
df = pd.read_csv('../data/raw/data_set_grupo_a.csv')

# 1. Informações básicas do dataset
print("=== Informações do Dataset ===")
print(df.info())

# 2. Verificando valores ausentes
print("\n=== Análise de Valores Ausentes ===")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

# 3. Estatísticas descritivas
print("\n=== Estatísticas Descritivas ===")
print(df.describe())

# 4. Distribuição da variável alvo
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Bankrupt?')
plt.title('Distribuição de Empresas Falidas vs Não Falidas')
plt.show()

# 5. Matriz de correlação
plt.figure(figsize=(15, 12))
correlation_matrix = df.corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, 
            mask=mask,
            annot=False,
            cmap='coolwarm',
            center=0)
plt.title('Matriz de Correlação')
plt.tight_layout()
plt.show()

# 6. Top 10 features mais correlacionadas com a falência
correlations = df.corr()['Bankrupt?'].sort_values(ascending=False)
print("\n=== Top 10 Features Mais Correlacionadas com Falência ===")
print(correlations[:10])

# 7. Distribuição das principais features
top_features = correlations[1:6].index
plt.figure(figsize=(15, 10))
for i, feature in enumerate(top_features, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='Bankrupt?', y=feature, data=df)
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 8. Salvando informações importantes
analise = {
    'missing_values': missing_values.to_dict(),
    'correlations': correlations.to_dict(),
    'descriptive_stats': df.describe().to_dict()
}

# Salvando em formato JSON para uso posterior
import json
with open('../data/processed/analise_exploratoria.json', 'w') as f:
    json.dump(analise, f)

print("\nAnálise exploratória concluída e salva com sucesso!")
```

```python
# Importando bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Configurando o estilo dos gráficos - usando um estilo padrão do matplotlib
plt.style.use('default')
sns.set_theme()  # Configuração do seaborn
plt.rcParams['figure.figsize'] = (12, 8)

# Carregando os dados
try:
    df = pd.read_csv('../data/raw/data_set_grupo_a.csv')
    
    # 1. Informações básicas do dataset
    print("=== Informações do Dataset ===")
    print(df.info())

    # 2. Verificando valores ausentes
    print("\n=== Análise de Valores Ausentes ===")
    missing_values = df.isnull().sum()
    print(missing_values[missing_values > 0])

    # 3. Estatísticas descritivas
    print("\n=== Estatísticas Descritivas ===")
    print(df.describe())

    # 4. Distribuição da variável alvo
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x='Bankrupt?')
    plt.title('Distribuição de Empresas Falidas vs Não Falidas')
    plt.show()

    # 5. Matriz de correlação
    plt.figure(figsize=(15, 12))
    correlation_matrix = df.corr()
    mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
    sns.heatmap(correlation_matrix, 
                mask=mask,
                annot=False,
                cmap='coolwarm',
                center=0)
    plt.title('Matriz de Correlação')
    plt.tight_layout()
    plt.show()

    # 6. Top 10 features mais correlacionadas com a falência
    correlations = df.corr()['Bankrupt?'].sort_values(ascending=False)
    print("\n=== Top 10 Features Mais Correlacionadas com Falência ===")
    print(correlations[:10])

    # 7. Distribuição das principais features
    top_features = correlations[1:6].index
    plt.figure(figsize=(15, 10))
    for i, feature in enumerate(top_features, 1):
        plt.subplot(2, 3, i)
        sns.boxplot(x='Bankrupt?', y=feature, data=df)
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # 8. Salvando informações importantes
    analise = {
        'missing_values': missing_values.to_dict(),
        'correlations': correlations.to_dict(),
        'descriptive_stats': df.describe().to_dict()
    }

    # Criando diretório se não existir
    import os
    os.makedirs('../data/processed', exist_ok=True)

    # Salvando em formato JSON para uso posterior
    import json
    with open('../data/processed/analise_exploratoria.json', 'w') as f:
        json.dump(analise, f)

    print("\nAnálise exploratória concluída e salva com sucesso!")

except FileNotFoundError:
    print("Erro: Arquivo não encontrado. Verifique se o caminho do arquivo CSV está correto.")
except Exception as e:
    print(f"Erro inesperado: {str(e)}")
```
